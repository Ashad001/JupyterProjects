```python
import random
name = input("Enter your name: ")
print(f"Hi, {name}! Welcome To Guessing Game!")
number = random.randint(1,10000)
play = 'yes'
count = 0

def Reset():
  global play
  play = input('Do You Want To Play Again? (Yes/No)')
  if play == 'yes' or play == 'Yes':
    global count, number
    number = random.randint(1,10000)
    play = 'yes'
    count = 0

while play == 'yes':
  x = int(input("Enter your guess: "))
  count += 1
  if x == number:
    print(f"\nGreat! You Found The Number In {count} moves.")
    Reset()
  elif x > number:
    print(f'\nYour Guess Is Higher, try some small numbers!')
    print(f"Moves: {count}")
  else:
    print(f"\nYour Guess is very low, try some higher value!")
    print(f"Moves: {count}")

```
